```python
import numpy as np
import pandas as pd
import regex as re
import os
```

```python
def find_questions_and_answers(s: string, i_:int):

    # The regular expression pattern to match the questions and answers
    pattern = r'Q:\s*(.*?)(?=(?:A:|Q:|\s*<h3>))A:\s*(.*?)(?=(?:Q:|\s*<h3>))'

    # Find all matches of the pattern in the string
    matches = re.findall(pattern, s, re.DOTALL)
    questions = []
    answers = []
    if len(matches) >= 1:
        print(i)
    # Loop through the matches and extract the questions and answers
    for match in matches:
        question = match[0].strip()
        answer = match[1].strip()
        questions += [question]
        answers += [answer]
        print("Question:", question)
        print("Answer:", answer)
    return questions, answers
```

```python
# Define a callback function to keep the contents of <h3> tags
def keep_h3(match):
    return match.group(0)
```

```python
directory = "downloaded"
i = 0
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        with open(f, "r") as read_file:
            content = read_file.read()

        # print(content)
        
        # Remove all HTML tags except for <h3> tags
        clean_html = re.sub(r'(?!<h3\b)\s*<.*?>\s*(?!</h3>)', '', str(content), flags=re.DOTALL)
        clean_html = re.sub(r'<h3\b.*?</h3>', keep_h3, clean_html, flags=re.DOTALL)
        
        #print(clean_html)
        questions, answers = find_questions_and_answers(clean_html, i)
        
        #remove all the html tags
        content2 = re.sub('<[^<]+?>', '', str(content))
        # print(content2)

        #remove all the \n
        content3 = re.sub(r'\\n', '', str(content2))
        # print(content3)

        #remove all the duplicate spaces
        content4 = re.sub(r'\s+', ' ', str(content3))
        # print(content4)

        #extract the questions<
        possible_ends = ["Guided Practice", "Explore More", "Practice","Guided Practice", "Review"]
        #find the first occurence of any of the possible ends
        occ = [content4.find(end) for end in possible_ends]
        idx = np.argmin([999999999 if o == -1 else o for o in occ])
        #split the content to separate the context from the questions
        # print(possible_ends[idx])
        strings = content4.split(possible_ends[idx])
        # print(strings[1])

        #save it
        np.savez(f"processed/lecture{i}.npz", lecture=strings, q=questions, a=answers)
        i+=1
        print("processed " + str(i) + " files")


```
