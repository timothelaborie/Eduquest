```python
import numpy as np
import pandas as pd
import urllib.request
import json
import regex as re
```

```python
content = urllib.request.urlopen("https://opentextbc.ca/catalog/bccampus").read()
content
```

```python
#extract every link
links = re.findall(r'href="([^"]*)"', content.decode('utf-8'))[8:224+8]

#remove duplicates
links = list(set(links))

#extract names of books
names = [link.split('/')[-1] for link in links]

#obtain link to book
links = [link + '/open/download?type=xhtml' for link in links]

```

```python
#download the books
for (link,name) in zip(links,names):
    print(link)
    try:
        book = urllib.request.urlopen(link).read()
    except Exception as e:
        print(e)
        continue


    #save the html file
    filename = "downloaded/" + str(name) + ".html"
    with open(filename, "wb") as f:
        f.write(book)

    # break
```
